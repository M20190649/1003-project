```python
import preprocess
import random
import matplotlib.pyplot as plt
from collections import defaultdict
%matplotlib inline
from matplotlib import rc
rc('figure', figsize=(16, 8), max_open_warning=False)
```

```python
mydict = preprocess.create_paper_paper_dict()
len(mydict.keys())
```

```python
resultdict = defaultdict(int)
for key, value in mydict.items():
    resultdict[len(value)] += 1
```

```python
# Papers which is citing more than 1500?
keylist = sorted(resultdict)
plt.plot(keylist, [resultdict[i] for i in keylist])
```

```python
# Build a random subset of dictionary, where we only retain references to themselves
random_keys = random.sample(mydict.keys(),10000)
random_dict = dict()
for each in random_keys:
    newkey=[]
    for ref in mydict[each]:
        if ref in random_keys:
            newkey.append(ref)
    random_dict[each] = newkey
```

```python
# most papers have 0 reference between each other in random subset of size 10k
summary_dict = defaultdict(int)
for key, value in random_dict.items():
    summary_dict[len(value)] += 1
summary_dict
```
